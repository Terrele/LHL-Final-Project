```python
def initialize():
    info = {}
    info['blue_average_level'] = 0                #10 minutes
    info['blue_total_gold'] = 0                   #10 minutes
    info['blue_totalDamageDoneToChampions'] = 0   #10 minutes
    info['blue_totalDamageDone'] = 0              #10 minutes
    info['blue_minionsKilled'] = 0                #10 minutes
    info['blue_timeEnemySpentControlled'] = 0     #10 minutes
    info['blue_jungleMinionsKilled'] = 0          #10 minutes
    info['blue_currentGold'] = 0                  #10 minutes
    info['blue_first_blood'] = 0
    info['blue_champions_killed'] = 0
    info['red_average_level'] = 0                 #10 minutes
    info['red_total_gold'] = 0                    #10 minutes
    info['red_totalDamageDoneToChampions'] = 0    #10 minutes
    info['red_totalDamageDone'] = 0               #10 minutes
    info['red_minionsKilled'] = 0                 #10 minutes
    info['red_timeEnemySpentControlled'] = 0      #10 minutes
    info['red_jungleMinionsKilled'] = 0           #10 minutes
    info['red_currentGold'] = 0                   #10 minutes
    info['red_first_blood'] = 0
    info['red_champions_killed'] = 0
    info['blue_win'] = None
    return info
```

```python
def add_kills(info):
    for i in range(11):
        try:
            for j in range(len(res.json()['info']['frames'][i]['events'])):
                if(res.json()['info']['frames'][i]['events'][j]['type'] == 'CHAMPION_KILL'):
                    player_kill_id = res.json()['info']['frames'][i]['events'][j]['killerId']
                    if player_kill_id in [1,2,3,4,5]:
                        info['blue_champions_killed'] += 1
                    else:
                        info['red_champions_killed'] += 1
                if(res.json()['info']['frames'][i]['events'][j]['type'] == 'CHAMPION_SPECIAL_KILL'):
                    if(res.json()['info']['frames'][i]['events'][j]['killType'] == 'KILL_FIRST_BLOOD'):
                        first_blood_id = res.json()['info']['frames'][i]['events'][j]['killerId']
                        if first_blood_id in [1,2,3,4,5]:
                            info['blue_first_blood'] += 1
                        else:
                            info['red_first_blood'] += 1
        except:
            break
    return info
```

```python
def add_stats(info):
    # try getting info 10 minutes into the game
    k = 10
    try:
        res.json()['info']['frames'][k]
    except:
        k = -1
    for i in range(1, 11):
        if i < 6:
            info['blue_average_level'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['level']
            info['blue_total_gold'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['totalGold']
            info['blue_totalDamageDoneToChampions'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['damageStats']['totalDamageDoneToChampions']
            info['blue_totalDamageDone'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['damageStats']['totalDamageDone']
            info['blue_minionsKilled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['minionsKilled']
            info['blue_timeEnemySpentControlled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['timeEnemySpentControlled']
            info['blue_jungleMinionsKilled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['jungleMinionsKilled']
            info['blue_currentGold'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['currentGold']
        else:
            info['red_average_level'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['level']
            info['red_total_gold'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['totalGold']
            info['red_totalDamageDoneToChampions'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['damageStats']['totalDamageDoneToChampions']
            info['red_totalDamageDone'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['damageStats']['totalDamageDone']
            info['red_minionsKilled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['minionsKilled']
            info['red_timeEnemySpentControlled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['timeEnemySpentControlled']
            info['red_jungleMinionsKilled'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['jungleMinionsKilled']
            info['red_currentGold'] += res.json()['info']['frames'][k]['participantFrames'][str(i)]['currentGold']
    info['blue_average_level'] /= 5
    info['red_average_level'] /= 5
    return info
```

```python
def winning_team(info):
    if(res.json()['info']['frames'][-1]['events'][-1]['winningTeam'] == 200):
        info['blue_win'] = 1
    else:
        info['blue_win'] = 0
    return info
```

```python
def match_data():
    info = initialize()
    info = add_kills(info)
    info = add_stats(info)
    info = winning_team(info)
    return info
```
